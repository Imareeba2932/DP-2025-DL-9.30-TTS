```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Flatten
```

```python
sentences = [
    "i love deep learning",
    "i love machine learning",
    "i love artificial intelligence"
]
```

```python
#tokenization
tokenizer = tf.keras.preprocessing.text.Tokenizer()
tokenizer.fit_on_texts(sentences)

sequences = tokenizer.texts_to_sequences(sentences)
word_index = tokenizer.word_index
```

```python
X = []
y = []

for seq in sequences:
    X.append(seq[:3])  # first three words as input
    y.append(seq[3])   # fourth word as output

X = np.array(X)
y = np.array(y)

vocab_size = len(word_index) + 1  # +1 for padding (if any)
```

```python
#Build LSTM model
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=8, input_length=3),
    LSTM(32), #number of neurons
    Dense(vocab_size, activation='softmax')
])
```

```python
#compiling model
model.compile(
    optimizer = 'adam',
    loss ='sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
```

```python
#training
model.fit(X,y,  epochs = 300)
#epochs 300 for better accuracy, verbose 0 is used to suppress output
```

```python
#test the model
test_text = ["i love deep"]
test_seq = tokenizer.texts_to_sequences(test_text)
test_seq = np.array(test_seq)
```

```python
prediction = model.predict(test_seq)
predicted_word_index = np.argmax(prediction, axis=1)[0]
```

```python
print("Input:", test_text[0])
print("predicted next word:", tokenizer.index_word[predicted_word_index])
```
